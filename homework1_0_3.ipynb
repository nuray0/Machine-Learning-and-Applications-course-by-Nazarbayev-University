```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```
#To generate 100 dataset
d = 2 #dimension is 2
w = np.random.randint(1,10, d) #weights
b = np.random.randint(1,10)  #some b
x = np.random.uniform(-1,1,(100,d))*10 #our inputs
y = np.zeros(100)  #corresponding outputs

x1 = np.random.uniform(-1,1,(1000,d))*10 #our inputs
y1 = np.zeros(1000)  #corresponding outputs

#some random line
h = x.dot(w) + b 
h1 = x1.dot(w) + b 
#and labels
label = (h > 0)*1
y = label
label1 = (h1 > 0)*1
y1 = label1
#plotting
plt.scatter(x[:, 0], x[:, 1], c=label)
plt.plot(x[:,0],-w[0]*x[:,0]/w[1]-b/w[1])
plt.show()
plt.scatter(x1[:, 0], x1[:, 1], c=label1)
plt.show()
```

```
# STEP 2 - PLA
#used Perceptron from the scikit-learn
perc = Perceptron()
perc.fit(x, y)

# predicted coefficients, w
wa1 = perc.coef_[0]
# Intercept or bias (constant), w0
wa0 = perc.intercept_

# Plotting the decision boundary
fig = plt.figure(figsize=(10,7))
#from min horizontal axis x value -5 to max + 5
x_values = [np.min(x[:, 1] -5 ), np.max(x[:, 1] +5 )]
# calculating y values from the weight update formula of PLA, line equation
y_values = np.dot((-1./wa1[1]), (np.dot(wa1[0],x_values) + wa0))
colors=['black' if l==0 else 'yellow' for l in y]
#plotting the dataset with color labels
plt.scatter(x[:, 0], x[:, 1], label='Perceptron Learning Algorithm', color=colors)
#plotting the line we have found using coefficients of PLA for each data sample
plt.plot(x_values, y_values, label='Decision Boundary')
plt.show()
```
